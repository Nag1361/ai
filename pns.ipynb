```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_data(file_path):
  pd.read_csv(file_path)
  return
```

```python
file_path = 'titanic_dataset_250-7.csv'
df = load_data(file_path)
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'].dropna(), bins=20, kde=True)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Distribution of Age among Passengers")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, alpha=0.7)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Fare vs Age with Survival Status")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Passenger ID', data=df, palette="pastel")  # Corrected to 'Pclass'
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.title("Number of Passengers per Class")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Passenger ID', hue='Survived', data=df, palette="muted")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.title("Survivors vs Non-Survivors per Class")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```
