```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
mean = 50
std_dev = 5
sample_size = 1000

# Generate random data points following a normal distribution
data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

# Plot Histogram
plt.figure(figsize=(10,6))
sns.histplot(data, kde=False, bins=30, color='skyblue', stat='density')

# Plot PDF (theoretical normal distribution curve)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-(x - mean)*2 / (2 * std_dev*2))
plt.plot(x, pdf, 'r', linewidth=2)

plt.title("Normal Distribution: Mean = 50, Std Dev = 5")
plt.xlabel("Data Points")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
import scipy.stats as stats

# Parameters
mean_test = 14
std_dev_test = 2.5

# I. Probability between 14 and 2.5
z_14 = (14 - mean_test) / std_dev_test
z_2_5 = (2.5 - mean_test) / std_dev_test
prob_14_2_5 = stats.norm.cdf(z_14) - stats.norm.cdf(z_2_5)

# II. Probability below 12
z_12 = (12 - mean_test) / std_dev_test
prob_below_12 = stats.norm.cdf(z_12)

# III. Probability above 19
z_19 = (19 - mean_test) / std_dev_test
prob_above_19 = 1 - stats.norm.cdf(z_19)

# Visualizing the normal distribution with probabilities

x = np.linspace(mean_test - 4*std_dev_test, mean_test + 4*std_dev_test, 1000)
pdf = stats.norm.pdf(x, mean_test, std_dev_test)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label="Normal Distribution", color="blue")
plt.fill_between(x, pdf, where=(x >= 12) & (x <= 14), color="green", alpha=0.5, label="Probability between 14 and 2.5")
plt.fill_between(x, pdf, where=(x <= 12), color="yellow", alpha=0.5, label="Probability below 12")
plt.fill_between(x, pdf, where=(x >= 19), color="red", alpha=0.5, label="Probability above 19")

plt.title("Normal Distribution for Test Scores (Mean=14, Std Dev=2.5)")
plt.xlabel("Test Score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()

# Output probabilities
print(f"Probability of scoring between 14 and 2.5: {prob_14_2_5:.4f}")
print(f"Probability of scoring below 12: {prob_below_12:.4f}")
print(f"Probability of scoring above 19: {prob_above_19:.4f}")
```
