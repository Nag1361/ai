```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file (replace 'your_file.xlsx' with the actual file name)
data = pd.read_excel('Real estate valuation data set.xlsx')

# Display the first few rows to verify the data
print(data.head())
print(data.columns)  # Print the column names to check for 'house_age'

# Assuming 'X2 house age' is the actual column name for house age
top_5_house_age = data.groupby('X2 house age')['Y house price of unit area'].mean().nlargest(5)
print("Top 5 house ages with the highest estate values:\n", top_5_house_age)

# Plotting the top 5 house ages with the highest estate value
plt.figure(figsize=(10, 6))
top_5_house_age.plot(kind='bar', color='teal')
plt.title('Top 5 House Ages with Highest Estate Value')
plt.xlabel('House Age (years)')
plt.ylabel('Average House Price per Unit Area')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Excel file (replace 'your_file.xlsx' with the actual file name)
data = pd.read_excel('Real estate valuation data set.xlsx')

# Print the actual column names from your DataFrame to identify the correct ones
print(data.columns)

# Correlation analysis using the actual column names from the DataFrame
# Replace the column names below with the correct ones from the output above
correlation_matrix = data[['Y house price of unit area', 'X3 distance to the nearest MRT station', 'X4 number of convenience stores']].corr()
print("\nCorrelation matrix:\n", correlation_matrix)

# Visualizing the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Real Estate Value')
plt.show()
```

```python
# Find the maximum and minimum house prices per unit area
# The actual column name in the dataset is 'Y house price of unit area'
max_price = data['Y house price of unit area'].max()
min_price = data['Y house price of unit area'].min()

print(f"Maximum house price per unit area: {max_price}")
print(f"Minimum house price per unit area: {min_price}")
```

```python
# Ensure 'transaction_date' is in datetime format
# Assuming the correct column name is 'X1 transaction date' based on Global Variables
data['X1 transaction date'] = pd.to_datetime(data['X1 transaction date'])

# Find the date with the most transactions
most_transactions_date = data['X1 transaction date'].mode()[0]
print(f"Date with the most real estate transactions: {most_transactions_date.date()}")
```

```python
# Define high-value properties as those in the top 10% of house price per unit area
# Changed 'house_price_per_unit_area' to the actual column name 'Y house price of unit area'
high_value_threshold = data['Y house price of unit area'].quantile(0.9)
high_value_data = data[data['Y house price of unit area'] >= high_value_threshold]

# Identify the closest distance perimeter to MRT stations for high-value properties
# Changed 'distance_to_mrt' to the actual column name 'X3 distance to the nearest MRT station'
min_distance_to_mrt = high_value_data['X3 distance to the nearest MRT station'].min()
max_distance_to_mrt = high_value_data['X3 distance to the nearest MRT station'].max()

print(f"Closest distance perimeter to MRT Stations for high real estate value:")
print(f"Min distance: {min_distance_to_mrt} meters")
print(f"Max distance: {max_distance_to_mrt} meters")

# Visualizing distribution of distances for high-value properties
plt.figure(figsize=(10, 6))
# Changed 'distance_to_mrt' to the actual column name 'X3 distance to the nearest MRT station'
sns.histplot(high_value_data['X3 distance to the nearest MRT station'], bins=20, kde=True, color='purple')
plt.title('Distance to MRT for High Real Estate Values')
plt.xlabel('Distance to MRT (meters)')
plt.ylabel('Frequency')
plt.show()
```
